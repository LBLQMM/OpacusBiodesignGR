# Notebook for translating gene names and preparing input files for Eflux2

```python
import pandas as pd
```

```python
#load gene translation
filename = "../data/lpd-to-opag.csv"
cols = ["lpd",'opag']
df=pd.read_csv(filename)
```

Make dic from csv using pandas:

```python
dic = pd.Series.from_csv(filename, header=None,).to_dict()
```

Test that dictionary works on LPD gene names and gives of OPAG gene names:

```python
dic['Pd630_LPD00286']
```

```python
dic.update({'#VALUE!': '#VALUE!'})
```

```python
df0 = pd.read_csv('../data/lpd-genes-from-transcripts.csv')
df_translated=df0.replace(dic)
```

```python
df_translated.to_csv("../data/translation-done.csv")
```

Now lets make the transcript files for Eflux2

```python
df_transcripts = pd.read_csv('../data/all-transcript-data.csv')
```

Let's get an overview of our transcript data

```python
df_transcripts.describe()
```

```python
#save all the transcript files for input into Eflux2
for i in list(df_transcripts):
    print('Saving transcript data for ',i)
    df_index = df_translated.copy(deep=True)
    df_final = pd.concat([df_index,df_transcripts], axis=1, ignore_index=False)
    df_final[['LPD',i]].to_csv(i+"_transcripts_for_eflux2.csv", index=False, header=False)
```
