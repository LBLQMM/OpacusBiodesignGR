This notebook calculates the growth rate, yield coefficient, and substrate uptake rate of R. opacus cultures growing with phenol and glucose. These experiements were published in this paper: [Henson (2018)](https://www.sciencedirect.com/science/article/pii/S1096717618300910).

#### Henson data:
* WT 0.5 g/L Phenol (3 trials) (**WT-P**)
* PVHG6 0.5 g/L Phenol (3 trials) (**PVHG-P**)

#### Rhiannon 2018 data (New data for this paper):
* Metabolomics and OD data for WT Glucose (**WT-G**)

### Method: 
<ol>
<li>Setup imports</li>
<li>Load phenol time-course growth and consumption data</li>
<li>Calculate wildtype phenol growth parameters</li>
<li>Create blank growth parameter dataframe</li>
<li>Add WT phenol data to dataframe</li>
<li>Calculate PVHG6 phenol growth parameters</li>
<li>Add PVHG6 phenol data to dataframe</li>
<li>Load glucose time-course growth and consumption data</li>
<li>Calculate WT glucose growth parameters</li>
<li>Add glucose growth parameters to dataframe</li>
<li>Save calculated growth parameters</li>
</ol>

### 1. Setup imports

```python
import sys
import pandas as pd

SourceDir = '../src'
sys.path.append(SourceDir)
from growth_parameter_functions import get_average_growth_parameters
```

### 2. Load phenol time-course growth and consumption data

```python
# load phenol data from csv files (The csv files have same format as EDD data)
phenol_od_df = pd.read_csv('../data/growth_data/phenol_growth.csv')
phenol_substrate_df = pd.read_csv('../data/growth_data/phenol_consumption.csv')
```

### 3. Calculate wildtype phenol growth parameters

```python
[growth_rate_WT_P, 
 yield_coeff_WT_P, 
 substrate_uptake_rate_WT_P, 
 growth_rate_std_WT_P, 
 yield_coeff_std_WT_P, 
 substrate_uptake_rate_std_WT_P] = get_average_growth_parameters(
    phenol_od_df, 
    phenol_substrate_df, 
    'WT-P-R1', 
    'WT-P-R2', 
    'WT-P-R3', 
    94.11, 
    substrate='phenol', 
    max_time=40
)

# print the growth parameters
print(f'growth_rate = {growth_rate_WT_P:.3f} ± {growth_rate_std_WT_P:.3f} hr-1')
print(f'yield coefficient = {yield_coeff_WT_P:.3f} ± {yield_coeff_std_WT_P:.3f} g biomass / mmol substrate')
print(f'substrate consumption rate = {substrate_uptake_rate_WT_P:.3f} ± {substrate_uptake_rate_std_WT_P:.3f} mmol substrate/gram biomass * hr')
```

### 4. Create blank growth parameter dataframe
The dataframe will hold the data calculated in this notebook

```python
# create dataframe for all results:
columns = ['growth rate', 'yield coefficient',  'substrate consumption rate', 'growth rate std dev', 'yield coefficient std dev', 'substrate consumption rate std dev']
strains = ['WT-P', 'PVHG-P', 'WT-G']

growth_parameter_df = pd.DataFrame(index=strains, columns=columns, dtype=float)
growth_parameter_df.head()
```

### 5. Add WT phenol data to dataframe

```python
growth_parameter_df.at['WT-P',:]=[growth_rate_WT_P, 
                                    yield_coeff_WT_P, 
                                    substrate_uptake_rate_WT_P, 
                                    growth_rate_std_WT_P, 
                                    yield_coeff_std_WT_P, 
                                    substrate_uptake_rate_std_WT_P
                                ]
growth_parameter_df
```

### 6. Calculate PVHG6 phenol growth parameters

```python
[growth_rate_PVHG6_P, 
 yield_coeff_PVHG6_P, 
 substrate_uptake_rate_PVHG6_P, 
 growth_rate_std_PVHG6_P, 
 yield_coeff_std_PVHG6_P, 
 substrate_uptake_rate_std_PVHG6_P] = get_average_growth_parameters(
    phenol_od_df, 
    phenol_substrate_df, 
    'PVHG-P-R1', 
    'PVHG-P-R2', 
    'PVHG-P-R3', 
    94.11, 
    substrate='phenol', 
    max_time=40
)

# print the growth parameters
print(f'growth_rate = {growth_rate_PVHG6_P:.3f} ± {growth_rate_std_PVHG6_P:.3f} hr-1')
print(f'yield coefficient = {yield_coeff_PVHG6_P:.3f} ± {yield_coeff_std_PVHG6_P:.3f} g biomass / mmol substrate')
print(f'substrate consumption rate = {substrate_uptake_rate_PVHG6_P:.3f} ± {substrate_uptake_rate_std_PVHG6_P:.3f} mmol substrate/gram biomass * hr')
```

### 7. Add PVHG6 phenol data to dataframe

```python
growth_parameter_df.at['PVHG-P',:]=[growth_rate_PVHG6_P, 
                                    yield_coeff_PVHG6_P, 
                                    substrate_uptake_rate_PVHG6_P, 
                                    growth_rate_std_PVHG6_P, 
                                    yield_coeff_std_PVHG6_P, 
                                    substrate_uptake_rate_std_PVHG6_P
                                ]
growth_parameter_df
```

### 8. Load glucose time-course growth and consumption data

```python
# load glucose data from csv files (The csv files have same format as EDD data)
glucose_od_df = pd.read_csv('../data/growth_data/glucose_growth.csv')
glucose_substrate_df = pd.read_csv('../data/growth_data/glucose_consumption.csv')
```

### 9. Calculate WT glucose growth parameters

```python
[growth_rate_WT_G, 
 yield_coeff_WT_G, 
 substrate_uptake_rate_WT_G, 
 growth_rate_std_WT_G, 
 yield_coeff_std_WT_G, 
 substrate_uptake_rate_std_WT_G] = get_average_growth_parameters(
    glucose_od_df, 
    glucose_substrate_df, 
    'WT-R1', 
    'WT-R2', 
    'WT-R3', 
    180.16, 
    substrate='glucose', 
    max_time=12
)

# print the growth parameters
print(f'growth_rate = {growth_rate_WT_G:.3f} ± {growth_rate_std_WT_G:.3f} hr-1')
print(f'yield coefficient = {yield_coeff_WT_G:.3f} ± {yield_coeff_std_WT_G:.3f} g biomass / mmol substrate')
print(f'substrate consumption rate = {substrate_uptake_rate_WT_G:.3f} ± {substrate_uptake_rate_std_WT_G:.3f} mmol substrate/gram biomass * hr')
```

### 10. Add glucose growth parameters to dataframe

```python
growth_parameter_df.at['WT-G',:]=[  
                                    growth_rate_WT_G, 
                                    yield_coeff_WT_G, 
                                    substrate_uptake_rate_WT_G, 
                                    growth_rate_std_WT_G, 
                                    yield_coeff_std_WT_G, 
                                    substrate_uptake_rate_std_WT_G
                                 ]
growth_parameter_df
```

### 11. Save calculated growth parameters

```python
growth_parameter_df.to_csv('../data/growth_parameters/experimental_growth_parameters.csv')
```
